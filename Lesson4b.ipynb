Layers and Activations with torch.nn


```python
import torch
from torch import nn
```

Linear Regression

```python
class LinearRegression(nn.Module):
  def __init__(self):
    super().__init__()
    self.func = nn.Linear(1,1)
  def forward(self,x):
    return self.func(x)

clfL = LinearRegression()
print(clf)
```

```python
x = torch.randn(1000,1)
```

```python
x[0]
```

```python
clfL(x[0])
```

Perceptron

```python
class Perceptron(nn.Module):
  def __init__(self):
    super().__init__()
    self.func = nn.Linear(1,1)
    self.activation = nn.Sigmoid()
  def forward(self,x):
    return self.func(x)

clf = Perceptron()
print(clf)
```

```python
clf(x[0])
```

Multi Layer Perceptron

```python
class MLP(nn.Module):
  def __init__(self):
    super().__init__()
    self.layers = nn.Sequential(
        nn.Linear(1,3),
        nn.Sigmoid(),
        nn.Linear(3,5),
        nn.Sigmoid(),
        nn.Linear(5,1),
    )

  def forward(self,x):
    return self.layers(x)

clfm= MLP()
clfm
```

```python
clfm(x[0])
```
